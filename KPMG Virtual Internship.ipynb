```python
# Females and Males tend to generate equal revenues
gender_wise_rev = merged_data[merged_data.order_status=="Approved"][['gender','list_price']].groupby(['gender']).sum().reset_index()
plt.pie(x=gender_wise_rev.list_price, labels=gender_wise_rev.gender, radius=1.5, autopct='%0.f%%');
```

```python
# Top 5 product IDs generating highest revenues
merged_data[merged_data.order_status=="Approved"][['product_id','list_price']].groupby(['product_id']).sum().sort_values(by='list_price', ascending=False).head(5)
```

```python
# Onine orders and Offline orders are same in revenue
online_wise_rev = merged_data[merged_data.order_status=="Approved"][['online_order','list_price']].groupby(['online_order']).sum().reset_index()
plt.pie(x=online_wise_rev.list_price, labels=online_wise_rev.online_order, radius=1.5, autopct='%0.f%%');
```

```python
# Brand-wise revenue: Solex has the highest share
brand_wise_rev = merged_data[merged_data.order_status=="Approved"][['brand','list_price']].groupby(['brand']).sum().reset_index()
plt.pie(x=brand_wise_rev.list_price, labels=brand_wise_rev.brand, radius=1.5, autopct='%0.f%%');
```

```python
# Product Class-wise revenue: Medium has the highest share
pclass_wise_rev = merged_data[merged_data.order_status=="Approved"][['product_class','list_price']].groupby(['product_class']).sum().reset_index()
plt.pie(x=pclass_wise_rev.list_price, labels=pclass_wise_rev.product_class, radius=1.5, autopct='%0.f%%');
```

```python
# Product Size-wise revenue: Medium has the highest share
psize_wise_rev = merged_data[merged_data.order_status=="Approved"][['product_size','list_price']].groupby(['product_size']).sum().reset_index()
plt.pie(x=psize_wise_rev.list_price, labels=psize_wise_rev.product_size, radius=1.5, autopct='%0.f%%');
```

```python
# Job_industry_category-wise revenue: Manufacturing then Financial Services the highest
industry_wise_rev = merged_data[merged_data.order_status=="Approved"][['job_industry_category','list_price']].groupby(['job_industry_category']).sum().reset_index()
plt.pie(x=industry_wise_rev.list_price, labels=industry_wise_rev.job_industry_category, radius=1.5, autopct='%0.f%%');
```

```python
# Wealth_segment-wise revenue: Mass Customer is the highest contributor in revenue
wealth_wise_rev = merged_data[merged_data.order_status=="Approved"][['wealth_segment','list_price']].groupby(['wealth_segment']).sum().reset_index()
plt.pie(x=wealth_wise_rev.list_price, labels=wealth_wise_rev.wealth_segment, radius=1.5, autopct='%0.f%%');
```
